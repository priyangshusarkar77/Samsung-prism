Title: Adaptive Color Scheme

About the technique:
The main motive of using this technique is to make the model invariant to colours. The color distribution of the training dataset will clearly have an impact on what colours the model is going to look for. For example, if there are more number of orange cats in the dataset, then the model might believe that cats are only in orange. So, to address these kinds of issues we have adaptive color scheme. Here, we will analyze the color distribution of each image in the training using histogram and inverse it so that the colors which are not in the images pops up. Now, among these colours we will select the top 6 colours and apply that colour shade on the image to produce new image with different shade of color.

Importing the required libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import random
import pathlib as pl
```

```python
#Using this function, we can find top colors which are present in the least amount in the image
def top_shades(new_hist, n_shades=6):
    freq = new_hist[0]
    freq = list(freq[1:-1])
    indexes = []
    maxi_val = 0
    while len(indexes) < n_shades:
        maxi_val = 0
        for i in freq:
            if i > maxi_val and freq.index(i) not in indexes:
                maxi_val = i
        indexes.append(freq.index(maxi_val))
    shades = []
    for i in indexes:
        shades.append(new_hist[1][i])
    return shades
```

```python
img = cv.imread(r"D:\Samsung Prism\Datasets\Cat and Dog\Training\Cat\cat.300.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img = cv.resize(img, (300, 300))
```

```python
#Getting the histogram distribution of each color channel
hist0 = cv.calcHist([img],[0],None,[256],[0,256])
hist1 = cv.calcHist([img],[1],None,[256],[0,256])
hist2 = cv.calcHist([img],[2],None,[256],[0,256])
```

```python
plt.plot(hist0, color="b")
plt.plot(hist1, color="g")
plt.plot(hist2, color="r")
```

```python
#Inversing the histogram so that the least present colours will come to the surface
new_hist0 = 255-hist0
new_hist1 = 255-hist1
new_hist2 = 255-hist2
```

```python
#Zeroing any negative values due to the inversions
new_hist0[new_hist0 < 0] = 0
new_hist1[new_hist1 < 0] = 0
new_hist2[new_hist2 < 0] = 0
```

```python
plt.plot(new_hist0, color="b")
plt.plot(new_hist1, color="g")
plt.plot(new_hist2, color="r")
```

```python
new_hist0 = np.histogram(new_hist0, bins=50)
new_hist1 = np.histogram(new_hist1, bins=50)
new_hist2 = np.histogram(new_hist2, bins=50)
```

```python
#Find the top shades of rgb channels which are not in the image
b = top_shades(new_hist0, n_shades=6)
g = top_shades(new_hist1, n_shades=6)
r = top_shades(new_hist2, n_shades=6)
```

```python
#Applying the shades and generating the new data
for i in range(6):
    b_shade = np.zeros((300, 300, 3))
    #print(b[i])
    for j in range(300):
        for k in range(300):
            for m in range(3):
                if m == 0:
                    b_shade[j][k][m] = r[i]/255
                if m == 1:
                    b_shade[j][k][m] = g[i]/255
                if m == 2:
                    b_shade[j][k][m] = b[i]/255
    op = cv.addWeighted(img, 0.5, b_shade, 0.5, 0)
    for p in range(300):
        for q in range(300):
            for l in range(3):
                op[p][q][l] = int(op[p][q][l]*255)
    op = op.astype("uint8")
    #Getting the histogram distribution of each color channel
    hist0 = cv.calcHist([op],[0],None,[256],[0,256])
    hist1 = cv.calcHist([op],[1],None,[256],[0,256])
    hist2 = cv.calcHist([op],[2],None,[256],[0,256])
    plt.plot(hist0, color="b")
    plt.plot(hist1, color="g")
    plt.plot(hist2, color="r")
    plt.figure()
    
    
    plt.title(b[i])
    plt.imshow(op)
    plt.figure()
```

```python
# For batch execution, pass the path of input file and output file

# Using this function, we will get the required histogram distributions
def req_colors(img, n_shades):
    hist0 = cv.calcHist([img],[0],None,[256],[0,256]) #b
    hist1 = cv.calcHist([img],[1],None,[256],[0,256]) #g
    hist2 = cv.calcHist([img],[2],None,[256],[0,256]) #r
    new_hist0 = 255-hist0
    new_hist1 = 255-hist1
    new_hist2 = 255-hist2
    new_hist0[new_hist0 < 0] = 0
    new_hist1[new_hist1 < 0] = 0
    new_hist2[new_hist2 < 0] = 0
    new_hist0 = np.histogram(new_hist0, bins=50)
    new_hist1 = np.histogram(new_hist1, bins=50)
    new_hist2 = np.histogram(new_hist2, bins=50)
    b = top_shades(new_hist0, n_shades)
    g = top_shades(new_hist1, n_shades)
    r = top_shades(new_hist2, n_shades)
    return b, g, r

# Using this function, we can overlay the new shade onto the image
def color_stack(img, b, g, r):
    shade = np.zeros((img.shape[0], img.shape[1], img.shape[2]), dtype="uint8")
    for j in range(img.shape[0]):
        for k in range(img.shape[1]):
            for m in range(img.shape[2]):
                if m == 0:
                    shade[j][k][m] = b
                if m == 1:
                    shade[j][k][m] = g
                if m == 2:
                    shade[j][k][m] = r
    op = cv.addWeighted(img, 0.5, shade, 0.5, 0)
    return op

# Main function to call all the sub functions
def adaptive_color_scheme(img, n_shades):
    b, g, r = req_colors(img, n_shades)
    result = []
    for i in range(n_shades):
        op = color_stack(img, b[i], g[i], r[i])
        result.append(op)
    return result
```

```python
# For batch execution

input_file_path = r""
output_file_path = r""

pathlib_file = pl.Path(input_file_path)
all_img_path = list(pathlib_file.glob("*"))

counter = 0
for i in all_img_path:
    img = cv.imread(str(i))
    result = adaptive_color_scheme(img, 6)
    random_num = random.randint(0, len(result)-1)
    cv.imwrite(output_file_path + "\\" + "aug_acs_" + str(counter) + ".jpg", result[random_num])
    counter += 1
```
